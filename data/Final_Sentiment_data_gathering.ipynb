```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import textacy.resources
import nltk
```

```python
df2 = pd.read_csv('transcripts.csv')
rs = textacy.resources.DepecheMood(lang="en", word_rep="lemmapos")
transcript = list(df2['transcript'])
df2['transcript']
```

```python
sentiment = []
for i in range(0,len(transcript)):
    text = transcript[i]
    doc = textacy.make_spacy_doc(text, lang='en_core_web_sm')
    emo_dict = rs.get_emotional_valence(doc)
    emo_list = [emo_dict.get('AFRAID'),emo_dict.get('AMUSED'),emo_dict.get('ANGRY'),emo_dict.get('ANNOYED'),emo_dict.get('DONT_CARE'),emo_dict.get('HAPPY'),emo_dict.get('INSPIRED'),emo_dict.get('SAD')]
    sentiment.append(emo_list)
df = pd.DataFrame(data=sentiment,columns=['AFRAID','AMUSED','ANGRY','ANNOYED','DONT_CARE','HAPPY','INSPIRED','SAD'])
df
```

```python
df.to_csv('sentiment_score.csv',index=False)
```
